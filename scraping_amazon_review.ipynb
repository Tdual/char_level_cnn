````python
import requests
from bs4 import BeautifulSoup
import bottlenose
import time
import json
import csv
from tqdm import tqdm_notebook as tqdm
````

````python
def sc(category):
    count = 0
    for p in tqdm(range(1,6)):
        url = "http://www.amazon.co.jp/gp/bestsellers/{}".format(category)
        params={
            "pg": p
        }
        htmltext = requests.get(url, params=params).text
        soup = BeautifulSoup(htmltext,"lxml")
        for el in soup.findAll("div",{"class":"zg_itemRow"}):
            time.sleep(5)
            meta = el.find("div",{"class":"a-fixed-left-grid p13n-asin"}).attrs["data-p13n-asin-metadata"]
            asign = json.loads(meta)["asin"]
            print(asign)
        
            url ="https://www.amazon.co.jp/product-reviews/{}/ref=cm_cr_getr_d_show_all".format(asign)
            for page in range(1,6):
                time.sleep(5)
                params={
                    "pageNumber": page
                }
                htmltext = requests.get(url, params=params).text
                soup = BeautifulSoup(htmltext, "lxml")
                
                l = []
                for e in soup.findAll("div",{"class":"a-section review"}):
                    body = e.findAll("span",{"data-hook":"review-body"})[0].text
                    rating = e.findAll("i",{"data-hook":"review-star-rating"})[0].findAll("span",{"class":"a-icon-alt"})[0].text.split("うち")[1]
                    l.append((rating, body))
                count += len(l)
                print(count)
                filename = "./data/amazon_ja/review_{}.csv".format(category)
                with open(filename,"a") as f:
                    writer = csv.writer(f, lineterminator='\n') 
                    for t in l:
                        writer.writerow(t) 
````
